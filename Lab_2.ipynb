```python
import pandas as pd
import numpy as np

def load_data(file_path, sheet_name):
    """Load data from an Excel file."""
    return pd.read_excel(file_path, sheet_name=sheet_name)

def segregate_matrices(data):
    """Segregate data into matrices A and C."""
    A = data.iloc[:, :-1].values
    C = data.iloc[:, -1].values
    return A, C

def calculate_dimensionality(A):
    """Calculate the dimensionality of the vector space."""
    return A.shape[1]

def calculate_number_of_vectors(A):
    """Calculate the number of vectors in the vector space."""
    return A.shape[0]

def calculate_rank(A):
    """Calculate the rank of matrix A."""
    return np.linalg.matrix_rank(A)

def calculate_costs(A, C):
    """Calculate the costs using the pseudo-inverse of matrix A."""
    A_pseudo_inverse = np.linalg.pinv(A)
    return A_pseudo_inverse.dot(C)
if __name__ == "__main__":
    file_path = 'Lab Session Data.xlsx'
    sheet_name = 'purchase data'
    data = load_data(file_path, sheet_name)
    A, C = segregate_matrices(data)
    dimensionality = calculate_dimensionality(A)
    print(f"Dimensionality of the vector space: {dimensionality}")

    num_vectors = calculate_number_of_vectors(A)
    print(f"Number of vectors in this vector space: {num_vectors}")

    rank_A = calculate_rank(A)
    print(f"Rank of Matrix A: {rank_A}")

    costs = calculate_costs(A, C)
    print("Costs of each product:")
    print(costs)
```
